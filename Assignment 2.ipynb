Ms. Gabriel Williams is a botany professor at District College. One day, she asked her student Mickey to compute the average of all the plants with distinct heights in her greenhouse.

Formula used:



```python
def average(arr):
    distinct_heights = set(arr)
    sum_of_distinct_heights = sum(distinct_heights)
    count_of_distinct_heights = len(distinct_heights)
    
    if count_of_distinct_heights == 0:
        return 0.0 
    avg = sum_of_distinct_heights / count_of_distinct_heights
    return round(avg, 3)

if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().split()))
    result = average(arr)
    print(result)
```

You are given an integer,N . Your task is to print an alphabet rangoli of size N . (Rangoli is a form of Indian folk art based on creation of patterns.)

Different sizes of alphabet rangoli are shown below:


The center of the rangoli has the first alphabet letter a, and the boundary has the Nth alphabet letter (in alphabetical order).

Function Description
Complete the rangoli function in the editor below.
rangoli has the following parameters:
int size: the size of the rangoli
Returns
string: a single string made up of each of the lines of the rangoli separated by a newline character (\n)
Input Format
Only one line of input containing size , the size of the rangoli.


```python
import string
def print_rangoli(size):
    alpha = string.ascii_lowercase
    lines = []
    
    for i in range(size):
        left = '-'.join(alpha[size-1:i:-1])
        center = alpha[i]
        right = '-'.join(alpha[i+1:size])
        row = (left + '-' + center + '-' + right).strip('-')
        lines.append(row.center(4*size-3, '-'))
    
    print('\n'.join(lines[::-1] + lines[1:]))

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
def merge_the_tools(string, k):
    num_substrings = len(string) // k

    for i in range(num_substrings):
        # Extract the current substring t_i
        t_i = string[i * k : (i + 1) * k]

        # Initialize an empty string for u_i and a set to keep track of seen characters
        u_i = ""
        seen_chars = set()

        # Iterate through the characters of t_i
        for char in t_i:
            # If the character has not been seen before, append it to u_i and add to seen_chars
            if char not in seen_chars:
                u_i += char
                seen_chars.add(char)
        
        # Print the resulting u_i for the current substring
        print(u_i)

if __name__ == '__main__':
    string = input()
    k = int(input())
    merge_the_tools(string, k)
```

```python
from collections import Counter

def calculate_raghu_earnings():
    X = int(input())

    # the space-separated list of all shoe sizes in the shop
    shoe_sizes_list = list(map(int, input().split()))

    # Using Counter to store the counts of each shoe size
    shoe_inventory = Counter(shoe_sizes_list)
    N = int(input()) #number of customers

    total_earnings = 0

    for _ in range(N):
        desired_size, price = map(int, input().split())
        
        if shoe_inventory[desired_size] > 0:
            shoe_inventory[desired_size] -= 1 # If available, "sell" the shoe by decrementing its count
            total_earnings += price

    print(total_earnings)

if __name__ == '__main__':
    calculate_raghu_earnings()
```

```python
def handle_division():
    T = int(input())  

    for _ in range(T):
        try:
            a_str, b_str = input().split()
            a = int(a_str) 
            b = int(b_str)
            
            result = a // b 
            print(result)
        except ZeroDivisionError as e:
            print("Error Code: " + str(e)) # Print specific error message for ZeroDivisionError
        except ValueError as e:
            print("Error Code: " + str(e)) # Print specific error message for ValueError
        except Exception as e:
            # This is a general catch-all for any other unexpected errors,
            # though the problem specifically asks for ZeroDivisionError and ValueError.
            print("An unexpected error occurred: " + str(e))

if __name__ == '__main__':
    handle_division()
```

```python
n = int(input())

s = set(map(int, input().split()))

N = int(input())

for _ in range(N):
    command_line = input().split()
    command = command_line[0]

    if command == "pop":
        if s:
            
            smallest_element = sorted(list(s))[0]
            s.remove(smallest_element)
    elif command == "remove":
        value = int(command_line[1])
        try:
            s.remove(value)
        except KeyError:
            pass
    elif command == "discard":
        value = int(command_line[1])
        s.discard(value)

print(sum(s))
```
